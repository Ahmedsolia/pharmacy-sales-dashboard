```python
import pandas as pd
import re

# 1. Normalize Arabic characters (strip Hamzas and change Ta Marbuta)
arabic_mapping = str.maketrans('أإآةى', 'اااهي')
df['account_name'] = df['account_name'].str.translate(arabic_mapping)

# 2. Safe cleaning (remove strange characters and brackets)
safe_noise_pattern = '[^a-zA-Z0-9\\s\u0600-\u06FF]'
df['account_name'] = df['account_name'].str.replace(safe_noise_pattern, ' ', regex=True)

# 3. Remove single English letters (e.g., dangling 'D' or 'C')
df['account_name'] = df['account_name'].str.replace(r'\b[a-zA-Z]\b', ' ', regex=True)

# 4. Convert single 'ص' (Sad) or 'س' (Seen) at the beginning of the text to 'صيدليه' (Pharmacy)
df['account_name'] = df['account_name'].str.replace(r'^[صس]\s+', 'صيدليه ', regex=True)

# 5. Convert single 'ش' (Sheen) to 'شارع' (Street)
df['account_name'] = df['account_name'].str.replace(r'\bش\s+', 'شارع ', regex=True)

# 6. Merge extra spaces and trim edges (must be the final polishing step)
df['account_name'] = df['account_name'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
# 1. Character Unification
arabic_mapping = str.maketrans('أإآةى', 'اااهي')
df['account_address'] = df['account_address'].str.translate(arabic_mapping)

# 2. Safe Noise Removal (Keeping digits)
# Numbers (0-9) are included here so they won't be deleted, and we removed the 'r' for PyArrow compatibility
safe_noise_pattern = '[^a-zA-Z0-9\\s\u0600-\u06FF]'
df['account_address'] = df['account_address'].str.replace(safe_noise_pattern, ' ', regex=True)

# 3. Replace "ش" with "شارع" (Street)
# This code looks for the letter "ش" on its own (whether at the beginning or preceded by a space) and converts it to "شارع"
# This method ensures that the letter 'ش' inside any word (like "Shubra" or "Shams") remains unaffected
df['account_address'] = df['account_address'].str.replace(r'(^|\s)ش(?=\s|$)', r'\1شارع', regex=True)

# 4. Final Cleanup (Removing extra spaces)
df['account_address'] = df['account_address'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
import pandas as pd
from collections import Counter

# 1. Function to extract potential areas from name and address
def extract_potential_areas(row):
    potential_areas = []
    
    name = str(row['account_name'])
    address = str(row['account_address'])
    
    # A. Extract the last word from the address (often the area)
    if address != 'nan' and address.strip():
        words = address.split()
        if len(words) > 1:
            potential_areas.append(words[-1]) # Last word
            potential_areas.append(words[-2] + " " + words[-1]) # Last two words (e.g., "Dar El-Salam")
            
    # B. Extract the word after the owner's name in account_name
    # Assuming the name is often 3 words, the 4th or 5th word is the area
    if name != 'nan' and name.strip():
        words = name.split()
        # Ignore initial words (Pharmacy, Doctor's name) and numbers at the end
        filtered_words = [w for w in words if not w.isdigit() and w not in ['صيدليه', 'شارع', 'س']]
        if len(filtered_words) >= 3:
             potential_areas.append(filtered_words[-1])

    return potential_areas

# 2. Apply the function and aggregate all potential words
all_potential_areas = []
for index, row in df.iterrows():
    all_potential_areas.extend(extract_potential_areas(row))

# 3. Count words to see the most frequent ones (these are likely the actual areas)
area_counts = Counter(all_potential_areas)

# 4. Convert result to a DataFrame for organized display
areas_df = pd.DataFrame(area_counts.items(), columns=['Word', 'Frequency']).sort_values(by='Frequency', ascending=False)

# Display the top 30 most frequent words
print(areas_df.head(30))
```

```python
import pandas as pd
import re
import numpy as np

# Function to extract the code and clean the name
def extract_branch_code(name):
    name = str(name)
    
    # This Regex searches for any word at the end of the name (that contains digits)
    match = re.search(r'([^\s]*\d+[^\s]*)$', name)
    
    if match:
        code = match.group(1) # The code we found
        
        # Remove the code from the original name
        clean_name = name.replace(code, '').strip()
        
        # Additional cleaning: If the user wrote (Cash, Credit, Installment, Supply Collection) 
        # before the code, remove it as well to clean the name
        clean_name = re.sub(r'(خط|نقدي|اجل|تحصيل سبلاي)\s*$', '', clean_name).strip()
        
        return pd.Series([clean_name, code])
    
    # If no numbers are found at the end, return the original name and a Null code
    return pd.Series([name, np.nan])

# Apply the function to create two new columns (Cleaned name and Branch code)
df[['clean_account_name', 'branch_code']] = df['account_name'].apply(extract_branch_code)

# Display a sample of the result to verify
print(df[['account_name', 'clean_account_name', 'branch_code']].dropna(subset=['branch_code']).head(15))
```

```python
import re

# 1. Text cleaning function using standard Python (to bypass PyArrow compatibility issues)
def clean_product_name(text):
    if pd.isna(text):
        return text
    # Remove any strange symbols, keeping Arabic characters, English letters, numbers, and spaces
    cleaned_text = re.sub(r'[^\sa-zA-Z0-9\u0600-\u06FF]', ' ', str(text))
    # Standardize double spaces and trim outer spaces
    return re.sub(r'\s+', ' ', cleaned_text).strip()

# 2. Safely apply the cleaning function to the column
df['product_name'] = df['product_name'].apply(clean_product_name)

# 3. Intelligent function to select the "most frequent name" (Mode) for each product
def get_most_frequent_name(series):
    mode_values = series.mode()
    # Choose the name that appears most often; if none, choose the first encountered
    return mode_values[0] if not mode_values.empty else series.iloc[0]

# 4. Apply unification based on supplier and product ID
df['product_name'] = df.groupby(['supplier_id', 'prod_id'])['product_name'].transform(get_most_frequent_name)

# 5. Display a sample of the cleaned data for review
print(df[['supplier_id', 'prod_id', 'product_name']].drop_duplicates().sample(30))
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

print("🚀 Starting Entity Resolution process...")

# 1. Extract unique entities (to avoid unnecessary calculations on 10,000 rows)
# We take supplier, account code, name, address, and area
entities = df[['supplier_id', 'account_id', 'account_name', 'account_address', 'area']].drop_duplicates().reset_index(drop=True)

# 2. Prepare text for matching (merging name with address to increase matching accuracy)
entities['match_text'] = entities['account_name'].fillna('') + ' ' + entities['account_address'].fillna('')
# Standardize empty areas into one block to compare them with each other
entities['area_block'] = entities['area'].fillna('UNKNOWN')

# 3. Prepare TF-IDF (using the magical 'char_wb' to handle Arabic spelling errors)
# This splits words into characters (e.g., "شفا" and "شفاء" will appear very similar)
vectorizer = TfidfVectorizer(analyzer='char_wb', ngram_range=(2, 4))

# 4. Prepare a Graph network to link matched pharmacies
G = nx.Graph()
for idx in entities.index:
    G.add_node(idx) # Each unique pharmacy is a "node" in the network

# Required similarity ratio to merge two pharmacies (85% is an excellent, tested ratio)
SIMILARITY_THRESHOLD = 0.85 

# 5. Matching within each region (Blocking) to save processing time
blocks = entities['area_block'].unique()
for block in blocks:
    # Identify row indices for the current region
    block_indices = entities[entities['area_block'] == block].index
    if len(block_indices) < 2:
        continue # If the region has less than one pharmacy, skip it
    
    # Convert texts to numbers and calculate similarity ratio
    block_texts = entities.loc[block_indices, 'match_text']
    tfidf_matrix = vectorizer.fit_transform(block_texts)
    sim_matrix = cosine_similarity(tfidf_matrix)
    
    # Extract matches that passed the Threshold
    for i in range(len(block_indices)):
        for j in range(i + 1, len(block_indices)):
            if sim_matrix[i, j] >= SIMILARITY_THRESHOLD:
                # If they are similar, draw an "edge" between them in the network
                G.add_edge(block_indices[i], block_indices[j])

# 6. Extract unified entities (Connected Components)
# This function gathers all pharmacies linked together and assigns them a single global code
components = list(nx.connected_components(G))

global_mapping = {}
for global_id, component in enumerate(components, start=1):
    # Create a unified code (Example: GLB_0001)
    new_global_code = f"GLB_{global_id:04d}"
    for node_idx in component:
        global_mapping[node_idx] = new_global_code

# 7. Add the unified code to the entities table
entities['Global_Account_ID'] = entities.index.map(global_mapping)

# 8. Merge the unified code into your original table
df = df.merge(
    entities[['supplier_id', 'account_id', 'Global_Account_ID']], 
    on=['supplier_id', 'account_id'], 
    how='left'
)

# --- Success Statistics ---
original_count = len(entities)
unified_count = entities['Global_Account_ID'].nunique()
print(f"✅ Matching completed successfully!")
print(f"📊 Number of entities before merging (with duplication across suppliers): {original_count}")
print(f"🎯 Actual number of entities (Global IDs): {unified_count}")
print(f"🔥 Detected and merged {original_count - unified_count} duplicate entities across suppliers.")
```
